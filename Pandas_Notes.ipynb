**Series:** It is defined as a one dimensional array that is capable of storing various data types

```python
import pandas as pd
x = [1,2,3,4,5]
a = pd.Series(x,index=['a','b','c','d','e'],dtype = float)#index used to chnge the index
print(a) #dtype used to change the data type
print(type(a))#returns type of the value
print(a[1])#to get value
```

***Merge & Concat***

```python
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[15,16,17,18]})
pd.merge(var_1,var_2,on="A")#only works on commom data
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[15,16,17,18]})
pd.merge(var_1,var_2,how = "left")
```

```python
pd.merge(var_1,var_2,how = "right")
```

```python
pd.merge(var_1,var_2,how = "outer",indicator=True)
```

```python
var_3=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var_4=pd.DataFrame({"A":[1,2,3,5],"B":[15,16,17,18]})
pd.merge(var_3,var_4,left_index=True,right_index=True,suffixes=("name","class"))
```

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,12,13,14])
pd.concat([sr1,sr2])
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2],"C":[15,16]})
pd.concat([var_1,var_2],axis = 1,join="inner")
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],"C":[15,16,17,18]})
pd.concat([var_1,var_2],keys=["x","y"],axis = 1)
```

**Join**

```python
import pandas as pd
varrr1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
varrr2 = pd.DataFrame({"c":[10,20,30,40],"D":[11,22,33,44]})
varrr1.join(varrr2)
```

```python
varrr1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
varrr2 = pd.DataFrame({"C":[10,20],"D":[11,22]})
varrr1.join(varrr2)
```

```python
varrr2.join(varrr1)
```

```python
varrr2.join(varrr1,how = "left")
```

```python
varrr2.join(varrr1,how = "right")
```

```python
varrr2 = pd.DataFrame({"B":[10,20],"D":[11,22]})
varrr2.join(varrr1,how = "outer",lsuffix="_1",rsuffix="_2")
```
