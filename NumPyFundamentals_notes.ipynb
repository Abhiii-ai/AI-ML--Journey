```python
#ARRAY ATTRIBUTES
```

```python
a1 = np.arange(10)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
#ndim gives dimension
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
#shape gives rows and colomns
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
#size gives number of items
print(a1.size)
print(a2.size)
print(a3.size)
```

```python
#itmsize = occupies size in mem
print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```

```python
#dtype = data type of items
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

```python
#changing datatype
```

```python
#astype
a3.astype(np.int32)
```

```python
#array operations
```

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

```python
#scalar operation
#arithmetic
print(a1*2)
print(a1+2)
print(a1-2)
print(a1/2)
print(a1**2)
```

```python
#relational
print(a2>15)
print(a2==15)

```

```python
#vector operations
#arithmetic
print(a1*a2)
print(a1+a2)
print(a1/a2)
```

```python
#ARRAY FUNCTIONS
```

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100) #round = round offs
a1
```

```python
#max/min/sum/prod
#0 -> col and 1-> row
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))
print(np.max(a1,axis=1))
print(np.max(a1,axis=0))
```

```python
#mean/median/std/var
print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))
print(np.mean(a1,axis = 1))
```

```python
#trignometric functions
np.sin(a1)
```

```python
#dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)
np.dot(a2,a3)
```

```python
#log and exponents
print(np.log(a1))
print(np.exp(a1))
```

```python
#round/floor/ceil
a4=np.random.random((2,3))*100
print(a4)
print(np.round(a4))
print(np.floor(a4))
print(np.ceil(a4))
```

```python
#STACKING
```

```python
#horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

print(np.hstack((a4,a5,a4,a5)))
```

```python
#vertical stacking
print(np.vstack((a4,a5,a4,a5)))
```
