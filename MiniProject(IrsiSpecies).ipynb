```python
import numpy as np

data1 = np.genfromtxt("Iris.csv", delimiter=",", dtype=str)

print(data1)
```

```python
header = data1[0]
data = data1[1:]
print(header)
print(data)
```

```python
num_data = data[:,1:5]
num_data = num_data.astype(float)
print(num_data)
```

```python
print("rows:",num_data.shape[0])
print("col:",num_data.shape[1])
print("shape:",num_data.shape)
print("dimension:",num_data.ndim)
print("datatype:", num_data.dtype)
print("size:",num_data.size)

```

```python
print("mean:",np.mean(num_data,axis = 0))
print("median:",np.median(num_data,axis = 0))
print("standard deviation:",np.std(num_data,axis = 0))
print("variance:",np.var(num_data,axis = 0))
print("max:",np.max(num_data,axis = 0))
print("min:",np.min(num_data,axis = 0))
```

```python
b1 = num_data[:,0]> 5
print(num_data[b1])
print("num:",np.sum(b1) ) #sepal length greater than 5
```

```python
b2 = num_data[:,2]> 4
print(num_data[b2])#petal length greater than 4
print("num:",np.sum(b2) )
```

```python
b3=num_data[:,1]<3
print(num_data[b3])#sepal width smaller than 3
print("num:",np.sum(b3) )
```

```python
b4=num_data[:,3]>1.5
print(num_data[b4])#petal width greater than 1.5
print("num:",np.sum(b4) )
```

```python
b5 = num_data[:,0] > 5
b6 = num_data[:,3]>1
print(num_data[b5&b6])
print("num:",np.sum(b5&b6) )
```
