# **Part 9: Apply Function**

**28. Create a new column:**
**Flower Size**

**Rule:**

* **Petal Length > 4 → Large**
* **Otherwise → Small**

**Use apply().**

```python
def size(PetalLengthCm):
  if PetalLengthCm > 4:
    return "Large"
  else:
    return "Small"
csv["Flower Size"] = csv["PetalLengthCm"].apply(size)
csv
```

**29. Create another column:
Sepal Area**

**Formula:**

**SepalLengthCm × SepalWidthCm**

```python
csv["Sepal Area"] = csv.apply(lambda x: x["SepalLengthCm"] * x["SepalWidthCm"],axis = 1)
csv
```
