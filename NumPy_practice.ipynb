```python
1. Create a 5×5 matrix of zeros.
2. Create a 4×4 identity matrix.
3. Generate 20 random integers between 1 and 100.
4. Find the mean, median, and standard deviation.
5. Reshape a 1D array into a 3×4 matrix.
6. Select all values greater than 50.
7. Multiply two matrices.
8. Find the maximum value in each row.
9. Reverse an array.
10. Find unique elements in an array.
```

```python
import numpy as np
#1
a1 = np.zeros((5,5))
print(a1)
```

```python
#2
a2 = np.identity(4,dtype=int)
print(a2)
```

```python
#3
a3 = np.random.randint(1,101,20)
print(a3)
```

```python
#4
print(np.mean(a3))
print(np.median(a3))
print(np.std(a3))
```

```python
#5
a4 = np.random.randint(1,101,12)
print(a4)
print(a4.reshape(3,4))
```

```python
#6
print(a4[a4 > 50])
```

```python
#7
a5 = np.arange(12).reshape(4,3)
a6 = np.arange(12,24).reshape(3,4)
print(np.dot(a5,a6))
```

```python
#8
np.max(a5,axis = 1)
```

```python
#9
np.flip(a4)
```

```python
#10
arr = np.array([1, 2, 3, 2, 4, 1, 5, 3])
print(np.unique(arr))
```
