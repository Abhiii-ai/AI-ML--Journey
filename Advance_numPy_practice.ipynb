```python
import numpy as np
a = np.random.randint(1,101,size=(10,10))
a
```

```python
print("max:",np.max(a))
```

```python
print("min:",np.min(a))
```

```python
print("max at index:",np.unravel_index(np.argmax(a),a.shape))
```

```python
print("min at index:",np.unravel_index(np.argmin(a),a.shape))
```

```python
np.where(a>50,100,a)
```

```python
indices = np.where(a>70)
print(indices)
```
